##### Выведите информацию о рождаемости детей в США по штатам. Для сортировки данных используйте функции sotr_index() и sort_values()

```python
baby_names.groupby('State')['Count'].sum().sort_values()
```

Рождаемость по штатам очень сильно разнится. Самое большое различие где-то 5 млн.

```python
baby_names.groupby('State')['Count'].count().sort_index()
```

##### Выведите 5 наиболее популярных имен. 

```python
print(baby_names.groupby('Name')['Count'].sum().sort_values(ascending=False).head(5))
```

Самое популярное имя - Jacob.

##### Выведите 10 наименее популярных имен. 

```python
baby_names.groupby('Name')[['Count']].sum().sort_values('Count').head(10)
```

Наименьшее владельцев каких-либо имен - 5.

##### Выведите 5 наиболее популярных женских имен и 5 наиболее популярных мужских имен. 

```python
female_names = baby_names[baby_names['Gender'] == 'F']
female_names.groupby('Name')[['Count']].sum().sort_values('Count', ascending=False).head(5)
```

```python
male_names = baby_names[baby_names['Gender'] == 'M']
male_names.groupby('Name')[['Count']].sum().sort_values('Count', ascending=False).head(5)
```

Самое популярное мужское имя - Jacob, самое популярное женское имя - Emma
